penjelasan metric

**keasaman:**
* Fixed acidity: kadar zat asam bawaan dari anggur yang sukar menguap (dasar dari keasaman wine, cthnya: asam tartrat, asam malat, asam sitrat)
* Volatile acidity: kadar asam asetat (mirip seperti cuka) yang menambah wangi wine
* citric acid: kadar asam sitrat dalam sejumlah anggur, menambah kesegaran ataupun keasaman wine yang kurang berasa




**zat yang mengawetkan:**
<br>sulfur dioksida/sulfit adalah kandungan utama yang digunakan pabrik untuk mengawetkan wine dan menghambat pertumbuhan bakteri dan ragi liar
* free sulfur dioxide: kadar sulfit yang aktif
* total sulfur dioxide: total kadar sulfit aktif dan sulfit yang sudah terikat ke molekul lain seperti gula dan aldehida
* sulphates: produk samping hasil fermentasi (sulfat)

ciri fisik:
<br>
* Density: semakin tinggi alkohol setelah proses fermentasi maka kadar air menurun
* pH: ph rendah (2.9-3.3)menghasilkan rasa asam dan rasa tidak tertinggal di tenggorokan, umumnya ada pada anggur putih. ph tinggi (3.4-4.0) memiliki rasa yang lebih lembut, tetapi lebih rawan mengalami pertumbuhan bakteri, umumnya ada pada anggur merah.
alcohol: kadar alcohol dalam wine, semakin tinggi semakin meninggalkan rasa yang panas

```python
# encoding untuk quality > 6.5 menjadi angka 1 untuk good
df['quality'] = np.where(df['quality'] > 6.5, 1, 0)

display(df.head())
```

```python
X = df.drop('quality', axis=1)
y = df['quality']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```
